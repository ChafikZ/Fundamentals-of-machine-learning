### Preprocessing

```python
# Scale the data to be between -1 and 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = data.values
X = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(X)
```

```python
std_scale = StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

```python
# Mean and variance per columns (axis=0) before scaling
np.mean(X, axis=0), np.var(X, axis=0)
```

```python
# After scaling
# The mean is close to 0 (not exactly due to the limited precision of computers)
# and the variance is 1. Scaling the variance to 1 for every features allows us
# to give the same importance to the variations of each features, even though they
# may have different amplitudes in the original dataset.
np.mean(X_scaled, axis=0), np.var(X_scaled, axis=0)
```

### PCA with sklearn

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
from sklearn.decomposition import PCA
```

```python
# Fit PCA only on *active* variables (months)
pca = PCA().fit(X_scaled[:,:12])
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
# 99% of the variance is explained by the first two components
# Project and plot the data the on first two components
X_proj = pca.transform(X_scaled[:,:12])
```

```python
_, axes = plt.subplots(ncols=4, figsize=(16,4))
for i, (ax, col) in enumerate(zip(axes, ['lat', 'long', 'moy', 'amp'])):
    ax.scatter(X_proj[:,0], data[col])
    ax.set_title(f'1st component vs {col}')
```

```python
_, axes = plt.subplots(ncols=4, figsize=(16,4))
for i, (ax, col) in enumerate(zip(axes, ['lat', 'long', 'moy', 'amp'])):
    ax.scatter(X_proj[:,1], data[col])
    ax.set_title(f'2nd component vs {col}')
```

```python
# The first component is strongly positively correlated with the mean (moy)
# The second component is strongly negatively correlated with the amplitude (ampl)
```

```python
plt.scatter(X_proj[:,0], X_proj[:,1], c=data['moy'])
plt.xlabel('1st principal component')
plt.ylabel('2nd principal component')
plt.xlim(-7, 7)
plt.ylim(-5, 5)

for i in range(X_proj.shape[0]):
    plt.text(X_proj[i,0], X_proj[i,1], data.index[i])
```

```python
# We can visually distinguish 3 clusters:
# - City with a warm temperature on average (Toulouse, Bordeaux, Montpellier, ...)
# - City with a cold temperature on average, and a large thermal amplitude (Strasbourg, Grenoble, ...)
# - City with a cold temperature on average, and a small thermal amplitude (Rennes, Nantes, Brest, ...)
```

```python
# n_components x n_features
components = pca.components_
```

```python
for i, (x, y) in enumerate(zip(components[0,:], components[1,:])):
    plt.plot([0, x], [0, y], color='k')
    plt.text(x, y, data.columns[i])

plt.plot([-0.7, 0.7], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-0.7, 0.7], color='grey', ls='--')

plt.xlim(-0.7, 0.7)
plt.ylim(-0.7, 0.7);
```

```python
# Features which have similar weights in the principal components are correlated (cf. features correlation plot)
```

### PCA manually using the covariance matrix
